```python
from abc import ABC, abstractmethod

lista_de_nomes_usados = []

class Banco():
    def __init__(self, nome):
        self.nome = nome
        self.agencias = [123,321,555]
        self.clientes = []
        self.contas = []
    
    def novo_cliente(self, cliente):
        self.clientes.append(cliente)
        
    def nova_conta(self, conta):
        self.contas.append(conta)
        
    def checar(self, cliente):
        checa = None
        if cliente not in self.clientes:
            print('Cliente desconhecido')
            return False
        for conta in cliente.contas:
            if conta in self.contas and conta.agencia in self.agencias:
                checa = True
        if not checa:
            print('Conta desconhecida ou Agencia Inexistente')
            return False
        else:
            return True

class Conta():
    def __init__(self, agencia, numero, saldo):
        self.agencia = agencia
        self.numero = numero
        self.saldo = saldo
    
    def deposito(self, valor):
        self.saldo += valor
        print(f'Novo saldo = {self.saldo}')
        
    @abstractmethod
    def sacar(self):
        pass

class ContaCorrente(Conta):
    def __init__(self, agencia, numero, saldo):
        super().__init__(agencia, numero, saldo)
        self.limite = 200
        
    def sacar(self, valor):
        if valor <= self.saldo + self.limite:
            if valor <= self.saldo:
                self.saldo -= valor
                print(f'Sacou R${valor} e ficou com R${self.saldo}')
            else:
                sobra = valor - self.saldo
                self.saldo = 0
                self.limite -= sobra
                print(f'Sacou R${valor} e ficou com R${self.saldo} de saldo e '\
                      f'R${self.limite - sobra} de limite (cheque especial)')
        else:
            print('Voce não tem Saldo nem Limite pra essa operação')

class ContaPoupanca(Conta):
    def sacar(self, valor):
        if valor <= self.saldo:
            self.saldo -= valor
            print(f'Sacou R${valor} e ficou com R${self.saldo}')
        else:
            print('Voce não tem Saldo nem Limite pra essa operação')        

class Pessoa():
    def __init__(self, nome: str, idade: int):
        # O Underscore indica que o atributo apenas deve ser usado por subclasses, e não fora da classe
        # Se fosse __nome pra acessar seria self._Pessoa__nome
        self._nome = nome
        self._idade = idade
        global lista_de_nomes_usados
        lista_de_nomes_usados.append(nome)
        
    @property
    def idade(self): #Método Getter Chamado
        return self._idade
    
    @property
    def nome(self): #Método Getter Chamado
        return self._nome

    @nome.setter 
    def nome(self, novonome): #Método setter
        global lista_de_nomes_usados
        if novonome not in  lista_de_nomes_usados:
            self._nome = novonome
        else: 
            print(f'Já Existe o nome {novonome}')
            
class Cliente(Pessoa):
    def __init__(self, nome, idade):
        super().__init__(nome, idade)
        self.contas = []
        
    def add_conta(self, conta):
        self.contas.append(conta)
    
    def lista_contas(self):
        print(f'{self._nome} tem {len(self.contas)} contas')
        
santander = Banco('santander')

p1 = Pessoa('Gabriel', 23)
p1.nome = 'Gabriel'
c1 = Cliente(p1._nome, p1._idade)
santander.novo_cliente(c1)
c1.lista_contas()
cc = ContaCorrente(123,0,0)
santander.nova_conta(cc)
c1.add_conta(cc)
c1.lista_contas()
c1.contas[0].deposito(200)
c1.contas[0].sacar(399)
if santander.checar(c1):
    print('Tudo certo meu caro')
```
